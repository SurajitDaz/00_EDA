# <span style="color:Blue"> Fixing Encoding Issue </span>
<span style="color:blue">Loading File 1 in DataFrame</span>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
companies=pd.read_csv("companies.csv", encoding = "ISO-8859-1")
#companies=pd.read_csv("companies.csv", engine='python')
companies
```

<span style="color:blue">As the last few rows are not encoded properly, the column is treated separately.</span>

```python
companies['permalink'] = companies['permalink'].str.encode('ISO-8859-1').str.decode('ascii', 'ignore')
companies.tail(3)
```

### Table-1.1 #2-> <span style="color:green"> Finding Total number of Unique Rows for the "companies" </span> 

```python
#len(companies["permalink"].str.lower().str.strip("/organization/-").unique())
companies["permalink"]=companies["permalink"].str.lower()
print(len(companies["permalink"].str.lower().unique()))
```

<span style="color:blue">Loading File 2 into DataFrame</span>

```python
rounds2=pd.read_csv("rounds2.csv", encoding = "ISO-8859-1")
rounds2
```

<span style="color:blue">As the last few rows are not encoded properly, the column is treated separately.</span>

```python
rounds2['company_permalink'] = rounds2['company_permalink'].str.encode('ISO-8859-1').str.decode('ascii', 'ignore')
rounds2.tail(3)
```

### Table-1.1 #1-> <span style="color:green">Finding Unique Records in "rounds2"</span>

```python
#len(rounds2["company_permalink"].str.lower().str.strip("/organization/-").unique())
rounds2["company_permalink"]=rounds2["company_permalink"].str.lower()
print(len(rounds2["company_permalink"].str.lower().unique()))
```

### <span style="color:green">Checking Records that belongs to rounds2 but not in companies.</span>

```python
filt=~(companies["permalink"].isin(rounds2["company_permalink"]))
companies.loc[filt]
```

### Table-1.1 #5-> <span style="color:green">Merging Two DataFrames & Finding Total Records.</span>

```python
Master_DataFrame=pd.merge(rounds2, companies, how="inner", left_on="company_permalink", right_on="permalink")
print(Master_DataFrame.shape)
```

<span style="color:blue">Looking inside Master_DataFrame</span>

```python
Master_DataFrame
```

<span style="color:blue">Now as "company_permalink" and "permalink" are basically same "permalink" can be dropped</span>

```python
Master_DataFrame=Master_DataFrame.drop(["permalink"], axis=1)
Master_DataFrame.shape
```

# <span style="color:Blue"> Imputation of Null Values</span>
<span style="color:blue">Checking for Null Values</span>

```python
100*Master_DataFrame.isna().sum()/len(Master_DataFrame)
```

<span style="color:blue">As in this analysis "funding_round_code" will not impact any respect the column can be dorpped though it has 73.54% null values.
Considering the given business objectives, same treatment can be done for "founded_at", "home_url", "state_code", "region",
"city", that means instead of imputation, these columns can be dropped, and this practise would make the dataframe more compact and increase the efficiency.</span>

```python
#Droping Columns
Master_DataFrame=Master_DataFrame.drop(["funding_round_code","homepage_url","state_code","region","city","founded_at"], axis=1)
print(Master_DataFrame.head())
print(Master_DataFrame.shape)
```

```python
#Summing up the missing value for remaining column & its relative frequency.
100*Master_DataFrame.isnull().sum()/len(Master_DataFrame.index)
```

<span style="color:blue">Now "raised_amount_usd" has high percentage of missing vale but this column plays an important role in our analysis and 
"country_code" is required for country wise analysis and "category_list" will be merged with another DataFrame. So instead of removing null values should be filled with some appropriate values following the imputation technique involving mean, median, mode, Random Sample Imputation, Imputation with End Distribution etc.<span>

```python
#Looking at statistical summary for column "raised_amount_usd"
Master_DataFrame["raised_amount_usd"].describe()
```

<span style="color:blue">Now it is seen the sprade of data (watching min and max) is huge; and also mean and median are far apart, which indicates that the distribution is a much skewed one. So in this case, imputation with mean is not right the approach, it is seen from the below aproach that, considering meadian generates ambiguity in the given range of investment analysis. So, Random Sample Imputation  techquniqe provides better results though but it can not be assumed that the data are missing completely at random(MCAR). So we also tri with End Distribution Imputation. In this case, distortion is very less </span>

<span style="color:blue">checking Distribution before End Distribution Imputation</span>

```python
Master_DataFrame=Master_DataFrame[(~np.isnan(Master_DataFrame["raised_amount_usd"]))]
100*Master_DataFrame.isnull().sum()/len(Master_DataFrame.index)
```

<span style="color:blue">Inspection for Treating missing values in  "country_code" column</span>

```python
#Statistical Description of ""country_code"
Master_DataFrame["country_code"].describe()
```

```python
cntry_cd=Master_DataFrame["country_code"].astype("category")
#Displaying The relative Frequency of each country
100*Master_DataFrame["country_code"].value_counts()/len(Master_DataFrame.index)
```

<span style="color:blue">As 62% records contain USA which is very big comparing to others in "country_code" column, 
the missing values in this column is replaced with "USA"  </span>

```python
Master_DataFrame["country_code"]=Master_DataFrame["country_code"].fillna("USA")
100*Master_DataFrame.isnull().sum()/len(Master_DataFrame.index)
```

<span style="color:blue">Treating missing values for the column "category_list" </span>

```python
Master_DataFrame["category_list"]
```

```python
#checking mode value
ss=Master_DataFrame["category_list"].mode(dropna=True)
#ss.iloc[0]
# filling na with mode
Master_DataFrame["category_list"]=Master_DataFrame["category_list"].fillna(ss.iloc[0])
```

```python
#Checking null values
100*Master_DataFrame.isnull().sum()/len(Master_DataFrame.index)
```

# <span style="color:blue">Analysis 3: Sector Analysis</span>

<span style="color:blue">It is required to extract the main sector treating "category_list" as data are available in mixed format, e.g  Advertising|Mobile|Web Development|Wireless.</span>

```python
#Using country filter to make the dataset smaller
Master_DataFrame=Master_DataFrame[cuntry_flt]
#extracting primary sector of each category from category _list column
Master_DataFrame.loc[:,'new_category_list']=Master_DataFrame['category_list'].apply(lambda x: x.split('|')[0])
Master_DataFrame['new_category_list']
```

```python
#grouping category_list
Master_DataFrame.groupby(['new_category_list'])["raised_amount_usd"].agg(SDF='sum').sort_values(by='SDF', ascending=False)
```

```python
#Reading the mapping file
map=pd.read_csv("mapping.csv",sep=",")
print(map.head())
map.shape
```

```python
#checking null value in mapping dataset
map.isnull().sum()
```

<span style="color:blue">As there is only one null record, we can drop that</span>

```python
map.dropna(inplace=True)
```

```python
#Converting all in cetegory_list & new_cetegory_list into lowercase
map.category_list=map.category_list.str.lower()
Master_DataFrame.new_category_list=Master_DataFrame.new_category_list.str.lower()
```

```python
flt=map.category_list.isin(Master_DataFrame.new_category_list)
```

```python
#Merging mapping dataset with Master_DataFrame
main_sector=pd.merge(Master_DataFrame, map, how="inner", left_on="new_category_list", right_on="category_list")
#drop "category_list_x", "category_list_y"
main_sector.drop(["category_list_x", "category_list_y"],axis=1, inplace=True)
main_sector.info()
main_sector.columns
```

<span style="color:blue">Now the column "new_category_list" is represented as the wide format in mapping dataframe (e.g. Health, Manufacturing,
News, Search and Messaging etc.). They contains 1 or 0.So it is converted to long format.</span>

```python
value_vars=main_sector.columns[9:18]
id_vars=np.setdiff1d(main_sector.columns, value_vars)
print(value_vars)
print(id_vars)
```

```python
long_df=pd.melt(main_sector, id_vars=list(id_vars), value_vars=list(value_vars))
long_df.head()
```

```python
long_df=long_df[long_df['value']==1]
long_df=long_df.drop('value',axis=1)
print(long_df.head())
print(len(long_df))
```

```python
long_df=long_df.rename(columns={'variable':'sector'})
long_df.info()
```

```python
main_sector=long_df[(long_df['raised_amount_usd']>=500000) & (long_df['raised_amount_usd']<=1500000)]
main_sector.columns
```
